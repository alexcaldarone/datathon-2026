```python
import requests

response = requests.post(
    "http://127.0.0.1:8000/listings",
    json={"query": "2 bedroom apartment in Zurich under 2500 CHF", "limit": 5, "offset": 0},
)

response.raise_for_status()
data = response.json()

for listing in data["listings"]:
    l = listing["listing"]
    print(f"[{listing['score']:.2f}] {l['title']} — {l['city']} — CHF {l['price_chf']}")
```

# OpenSearch

```python
from opensearchpy import OpenSearch
import os
from dotenv import load_dotenv
load_dotenv()

auth = ('team4', 'cjarp*o84%75q9Q') # For testing only. Don't store credentials in code.
host = os.environ["OPENSEARCH_ENDPOINT"]
client = OpenSearch(
    hosts=[{"host": host, "port": 443}],
    http_auth=auth,
    http_compress=True,  # enables gzip compression for request bodies
    use_ssl=True,
    verify_certs=True,
    ssl_assert_hostname=False,
    ssl_show_warn=False,
)

```

```python
client.indices.exists(index="listings")
```

```python
index_name = 'python-test-index'
index_body = {
  'settings': {
    'index': {
      'number_of_shards': 4
    }
  }
}

response = client.indices.create(index=index_name, body=index_body)
```

```python
document = {
  'title': 'Moneyball',
  'director': 'Bennett Miller',
  'year': '2011'
}

response = client.index(
    index = 'python-test-index',
    body = document,
    id = '1',
    refresh = True
)
```
